```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV

from sklearn.tree import DecisionTreeClassifier, plot_tree

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

import matplotlib.pyplot as plt
import seaborn as sns
import pickle
```

```python
df = pd.read_csv("drug200.csv")
df
```

```python
df.info()
```

```python
df['Sex'].value_counts()
```

```python
df['Sex'].replace({"M":1,"F":0},inplace =True)
```

```python
df['BP'].value_counts()
```

```python
df['BP'].replace({"HIGH":3,"LOW":2,'NORMAL':1},inplace =True)
```

```python
df['Cholesterol'].value_counts()
```

```python
df['Cholesterol'].replace({"HIGH":1,'NORMAL':1},inplace =True)
```

```python
df.info()
```

```python
x = df.drop('Drug',axis=1)
y = df['Drug']

x_train,x_test, y_train, y_test = train_test_split(x,y,random_state=1, stratify=y)

dt_clf = DecisionTreeClassifier(criterion = 'gini',random_state=1)
dt_clf.fit(x_train, y_train)
```

```python
acc_train = dt_clf.score(x_train,y_train)
acc_train
```

```python
acc_test = dt_clf.score(x_test,y_test)
acc_test
```

```python
plt.figure(figsize = (100,90))
tree = plot_tree(dt_clf, feature_names = list(x.columns),class_names=['drugY','drugX','drugA','drugC','drugB'],filled=True)
plt.savefig('DT_clf.png')
```

## training data evaluation

```python
y_pred_train = dt_clf.predict(x_train)

cnf_matrix = confusion_matrix(y_pred_train,y_train)
print("Confusion matrix :\n",cnf_matrix)
print("*"*59)
Accuracy = accuracy_score(y_pred_train,y_train)
print("Accuracy score :",Accuracy)
print("*"*50)

clf_report = classification_report(y_pred_train,y_train)
print("Classififcation report :",clf_report)
```

## testing data evaluation

```python
y_pred = dt_clf.predict(x_test)

cnf_matrix = confusion_matrix(y_pred,y_test)
print("Confusion matrix :\n",cnf_matrix)
print("*"*59)
Accuracy = accuracy_score(y_pred,y_test)
print("Accuracy score :",Accuracy)
print("*"*50)

clf_report = classification_report(y_pred,y_test)
print("Classififcation report :",clf_report)
```

## Pickle file

```python
with open("Drug.pkl",'wb') as f:
    pickle.dump(dt_clf,f)
```
