```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV

from sklearn.tree import DecisionTreeClassifier, plot_tree

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('drug200.csv')
df
```

```python
### EDA
df.info()
```

```python
df.describe()
```

```python
df['Sex'].value_counts()
df['Sex']=df['Sex'].replace({'M':1,'F':0})
```

```python
df['BP'].value_counts()
df['BP'] = df['BP'].replace({'HIGH':3,'LOW':2,'NORMAL':1})
```

```python
df['Cholesterol'].value_counts()
df['Cholesterol']=df['Cholesterol'].replace({'HIGH':1,'NORMAL':0})
```

```python
df
```

```python
### Train model

x = df.drop('Drug',axis=1)
y = df['Drug']

x_train,x_test, y_train, y_test = train_test_split(x,y,random_state=1, stratify=y)

dt_clf = DecisionTreeClassifier(criterion = 'gini',random_state=1)
dt_clf.fit(x_train, y_train)
```

```python
acc_train = dt_clf.score(x_train,y_train)
acc_train
```

```python
acc_test = dt_clf.score(x_test,y_test)
acc_test
```

```python
x_train
```

```python
y_train.value_counts()
```

```python
plt.figure(figsize = (100,90))
tree = plot_tree(dt_clf, feature_names = list(x.columns),class_names=['drugY','drugX','drugA','drugC','drugB'],filled=True)
plt.savefig('DT_clf.png')
```

```python
## Training data evaluation

y_pred_train = dt_clf.predict(x_train)

cnf_matrix = confusion_matrix(y_pred_train,y_train)
print('confusion matrix \n',cnf_matrix)
clf_report = classification_report(y_pred_train,y_train)
print('\nclassification report \n',clf_report)
accuracy = accuracy_score(y_pred_train,y_train)
print('\nAccuracy',accuracy)
```

```python
## Training data evaluation

y_pred_test = dt_clf.predict(x_test)

cnf_matrix = confusion_matrix(y_pred_test,y_test)
print('confusion matrix \n',cnf_matrix)
clf_report = classification_report(y_pred_test,y_test)
print('\nclassification report \n',clf_report)
accuracy = accuracy_score(y_pred_test,y_test)
print('\nAccuracy',accuracy)
```

### We are getting 100% accuracy so no need to go for pruning and Hyperparameter tuning
